```python
import pandas as pd
import geopandas as gpd
import pyproj
```

```python
home_path = 'D:\\OneDrive\\Data Science\\Data\\CO2\\'
df_emissions = pd.read_csv(home_path + 'historical_emissions_climatewatchdata_tidy.csv', sep=';')

pyproj.datadir.set_data_dir('C:\\Users\\simon\\anaconda3\\envs\\geo_env\\Library\\share\\proj')
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
df_emissions = df_emissions.merge(world, left_on='Country', right_on='name', how='inner')
geo_df_emissions = gpd.GeoDataFrame(df_emissions)
geo_df_emissions_2018 = geo_df_emissions[(geo_df_emissions['Year']==2018)]
```

```python
geo_df_emissions_2018.plot(
    'Per_capita_emissions',
    legend='Per_capita_emissions',
    # tooltip=['Country', 'Per_capita_emissions'],
    # style_kwds=dict(color="black"),
    # tiles="CartoDB positron",
    cmap="Reds",
    figsize=(16, 10),
    aspect='auto'
    );
```
